**Entropy, Cross-Entropy and KL-Divergence**

**Entropy**, as it relates to machine learning, is a measure of the randomness in the information being processed. The higher the entropy, the harder it is to draw any conclusions from that information. 

**Cross entropy** is often used as a loss function in machine learning.

Cross entropy is a measure from the field of information theory that builds on entropy and generally calculates the difference between two probability distributions. It is closely related to, but different from, **KL divergence**, which calculates the relative entropy between two probability distributions, whereas cross entropy is intended to calculate the total entropy between distributions.

Those concepts come from Claude Shannon's Information Theory. In Information Theory, the goal is to reliably and efficiently transmit a message from a sender to a recipent. In our digital age, messages are composed of bits. A bit is a number that is either equal to 0 or 1. But not all bits are useful: some of them are redundant, some of them are errors etc. So, when we communicate a message, we want to transmit as much useful information as possible to the recipient. In **Claude Shannon**’s paper, “The Mathematical Theory of Communication(1948)”, he stated that "*to transmit 1 bit of information means to reduce the recipient’s uncertainty by a factor of 2.*"

For example, if we have the information that weather is going to be rainy tomorrow by 50% chance, which means it can be sunny or rainy with 50-50% chance. This information gives us 1 bit of information. However, if we have 8 equally options for the weather prediction, this means we have 3 bits of information since the 8 is 3th power of 2. We can compute it with binary logarithm of the uncertainty reduction factor.

If the possibilities are not equally likely, we need to calculate the average bits multiply by possibilities. For example, the weather is going to be sunny with the possibility of 75% and rainy with the possibility of 25%  tomorrow. By using inverse log function we can calculate number of bits of useful information. For 25% possibility, -log(0.25) = 2,- and for 75% possibility -log(0.75) = 0.41. 

When we multiply number of bits with possibilities we can compute the average number of bits of useful information. This number changes depending on the number of possibilities and how varies of the possibilities within the distribution.
$$-\sum_{i\ =\ 1}^{i=w}{p(i)\ *\log}_w{p(i)}$$ 

```python
#  number of bits of useful information for the weather condition

ave_bits = (0.75*0.41) + (0.25*2)
print(ave_bits)
```

In general, it gives the average amount of information that we get from one sample drawn from a given probability distribution p . It tells us how unpredictable the probability distribution is.

Cross-entropy is a measure from the field of information theory, building upon entropy and generally calculating the difference between two probability distributions. It is closely related to but is different from KL divergence that calculates the relative entropy between two probability distributions, whereas cross-entropy can be thought to calculate the total entropy between the distributions.

In mathematical statistics, the Kullback–Leibler (KL) divergence, Dkl (also called relative entropy), is a measure of how one probability distribution is different from a second, reference probability distribution. Applications include characterizing the relative (Shannon) entropy in information systems, randomness in continuous time-series, and information gain when comparing statistical models of inference.

For an example, we have cards with 4 different colors: red, white, green and grey. In the beginning, since we don't have any additional information about them all possibilities are equal for each color when we pull a card from a bag. However, if we have any additional information about the distribution of colors the entropy will be change. In this example, we have information that number of red cards is much higher than others. As we may seen in the below application, entropy for second distribution is higher than the first condition.

```python
from math import log2
from matplotlib import pyplot
import matplotlib

matplotlib.rcParams.update({'font.size': 15, 'font.family': 'sans-serif', 'mathtext.fontset': 'stix'})
fig, axes = pyplot.subplots(figsize=(8,8))
pyplot.style.use('seaborn-pastel')

# define distributions
events = ['red', 'white', 'green', 'grey']
p = [0.25, 0.25, 0.25, 0.25]
q = [0.25, 0.25, 0.25, 0.25]
print('P=%.2f Q=%.2f' % (sum(p), sum(q)))

# plot first distribution for equal possibilities
pyplot.subplot(2,1,1)
pyplot.scatter(events, p)
pyplot.title("First Distribution", color = 'r', loc='center')

# show the plot
pyplot.show()

def cross_entropy(p, q):
	return -sum([p[i]*log2(q[i]) for i in range(len(p))])
 
print(f"Entropy value: {cross_entropy(p,q)}")
```

```python
matplotlib.rcParams.update({'font.size': 15, 'font.family': 'sans-serif', 'mathtext.fontset': 'stix'})
fig, axes = pyplot.subplots(figsize=(8,8))
pyplot.style.use('seaborn-pastel')

# define distributions
events = ['red', 'white', 'green', 'grey']
p = [0.25, 0.25, 0.25, 0.25]
q = [0.6, 0.2, 0.085, 0.015]
print('P=%.2f Q=%.2f' % (sum(p), sum(q)))

# plot second distribution
pyplot.subplot(2,1,2)
pyplot.scatter(events, q)
pyplot.title("Second Distribution", color = 'r', loc='center')

# show the plot
pyplot.show()

def cross_entropy(p, q):
	return -sum([p[i]*log2(q[i]) for i in range(len(p))])
 
print(f"Entropy value: {cross_entropy(p,q)}")
```
